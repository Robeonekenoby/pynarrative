```python
import pynarrative as pn
import pandas as pd

# Dati di esempio
data = pd.DataFrame({
    'Anno': range(2018, 2023),
    'Vendite': [100, 120, 90, 150, 200]
})

# Creazione del grafico completo in un unico flusso
final_chart = (pn.Story(data, width=600, height=400)
    .mark_line(color='blue')
    .encode(x='Anno:O', y='Vendite:Q')
    .add_title("Trend Vendite", "2018-2022", title_color="#1a1a1a", subtitle_color="#4a4a4a")
    .add_context("Crescita costante", position='top', color="#2ecc71")
    .add_call_to_action("Clicca per dettagli", position='bottom', color="#3498db")
    .add_annotation(
        2020, 90, "Punto di interesse vn,gsdkgkjfhwthkjuqehfrkunb", 
        arrow_direction='left', arrow_dx=50, arrow_dy=-50, arrow_color='red', arrow_size=75,
        label_color='darkgreen', label_size=14,
        show_point=True  
    )
    .add_call_to_action(
    type='line_steps',
    texts=["Step 1", "Step 2", "hot dogs", "puzzona"],
    position='bottom'   
    )
    .add_call_to_action(
    type='button',
    text="eééhhh",
    url="https://example.com",
    position='top',    
    title="Prossimi passi"
    )
    .add_call_to_action(
    type='stair_steps',
    texts=["Step 1", "Step 2", "canini", "tondina", "Balena"],
    position='right' 
    )
    
    #.configure_view(fill='#f0f0f0')
    .render()
)

# Visualizzazione del grafico
final_chart
```

# Example with the latest changes

```python
import pynarrative as pn
import pandas as pd

# Dati di esempio
data = pd.DataFrame({
    'Anno': range(2018, 2023),
    'Vendite': [100, 120, 90, 150, 200]
})

# Creazione del grafico completo in un unico flusso
final_chart = (pn.Story(data, width=600, height=400)
    .mark_line(color='blue')
    .encode(x='Anno:O', y='Vendite:Q')
    .add_title("Trend Vendite", "2018-2022", title_color="#1a1a1a", subtitle_color="#4a4a4a")
    .add_context("Crescita costante", position='top', color="#2ecc71")
    .add_call_to_action("Clicca per dettagli", position='bottom', color="#3498db")
    .add_annotation(
        2020, 90, "Punto di interesse vn,gsdkgkjfhwthkjuqehfrkunb", 
        arrow_direction='left', arrow_dx=50, arrow_dy=-50, arrow_color='red', arrow_size=75,
        label_color='darkgreen', label_size=14,
        show_point=True  # Ora funziona anche con False
    )
    .add_call_to_action(
    type='line_steps',
    texts=["Step 1", "Step 2", "hot dogs", "hamburger"],
    position='bottom',   
    line_steps_chart_width=350,
    line_steps_chart_height=50,
    line_steps_font_size=6,
    )
    .add_call_to_action(
    type='button',
    text="eééhhh",
    url="https://example.com",
    position='top',    
    title="Prossimi passi",
    
    title_color='blue',
    title_font_family='Helvetica',
    title_font_size=26,
    button_width=150,
    button_height=40,
    button_color='#FF0000',
    button_opacity=0.3
    )
    .add_call_to_action(
    type='stair_steps',
    texts=["Step 1", "Step 2", "aaa", "bo"],
    position='right'  
    )
    
               
     # NextStep
    .add_call_to_action(
        hot_dogs=pn.NextStep(
            type='button',
            text='Click me',
            position= 'left',
            url='https://example.com'
    )
    )
          
               
               
    #.configure_view(fill='#f0f0f0')
    .render()
)



# Visualizzazione del grafico
final_chart
```
